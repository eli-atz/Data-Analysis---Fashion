# Users Analysis:

```python
# Count sales per user_uuid
sales_per_user = df.groupby('user_id').size().reset_index(name='sales_count')

# Sort users by sales count in descending order and take the top 10
top_10_selling_users = sales_per_user.sort_values(by='sales_count', ascending=False).head(10)

# Merge with original DataFrame to get nationality
top_10_selling_users_with_nationality = pd.merge(top_10_selling_users, df[['user_id', 'country']], on='user_id', how='left')

# Display all rows
pd.set_option('display.max_rows', None)

# Print the result
print("Top 10 selling users and their nationalities:")
print(top_10_selling_users_with_nationality)
```

I would have liked to know the nationality of the users who sell the most within the marketplace, but I noticed that the analysis indicates that multiple nationalities are correlated with each user ID. Shouldn't the user ID be unique, and therefore their nationality as well?

# Country Analysis:

```python
# Exploring the distribution by country
country_distribution = df['country'].value_counts().head(3)
print("Country distribution:\n", country_distribution)

plt.figure(figsize=(6, 3))
sns.countplot(x='country', data=df)
plt.title('Country distribution')
plt.show()
```

# Categories Analysis:

```python
# Calculation of the percentage of categories compared to items
category_percentage = df['category'].value_counts(normalize=True) * 100
print("Category Percentage:\n", category_percentage)

# Creation of the pie chart
plt.figure(figsize=(6, 6))
plt.pie(category_percentage, labels=category_percentage.index, autopct='%1.1f%%', startangle=140)
plt.title('Category Percentage')
plt.axis('equal')
plt.show()
```

```python
# Explore the most populated clothing categories
top_categories = df['category'].value_counts().head(10)
print("Category Distribution:\n", top_categories)

# Bar chart of the top categories
plt.figure(figsize=(6, 4))
sns.barplot(x=top_categories.values, y=top_categories.index, palette='viridis')
plt.title('Category Distribution')
plt.xlabel('Number of Items')
plt.ylabel('Category')
plt.show()
```

```python
plt.figure(figsize=(6, 3))
df.groupby('category')['stars'].mean().plot(kind='bar')
plt.title('Average ratings by category')
plt.xlabel('Category')
plt.ylabel('Average star rating')
plt.show()
```

The analysis has confirmed that the categories of belts and shoes are among the most appreciated by purchasers.

# Designers Analysis

```python
# Explore the most popular designers
top_designers = df['designer_id'].value_counts().head(10)
print("Top 10 Designers:\n", top_designers)

# Bar chart of the top designers
plt.figure(figsize=(8, 6))
sns.barplot(x=top_designers.index, y=top_designers.values, palette='viridis')
plt.title('Top 10 Designers')
plt.xlabel('Designer')
plt.ylabel('Number of Items')
plt.show()
```

```python
# Analysis of reviews for each designer
average_stars = df.groupby('designer_id')['stars'].mean()

# Bar chart of the average stars for each designer
plt.figure(figsize=(9, 4))
sns.barplot(x=average_stars.index, y=average_stars.values)
plt.title('Average stars for each designer')
plt.xlabel('Designer ID')
plt.ylabel('Average Stars')
plt.show()
```

```python
# Count sales per designer_id
sales_per_designer = df.groupby('designer_id').size().reset_index(name='sales_count')

# Sort designers by sales count in descending order
top_10_selling_designers = sales_per_designer.sort_values(by='sales_count', ascending=False).head(10)

# Merge with original DataFrame to get category and level
top_10_selling_designers_with_info = pd.merge(top_10_selling_designers, df[['designer_id', 'category', 'level']], on='designer_id', how='left')

# Display all rows
pd.set_option('display.max_rows', None)

# Print the result
print("Top 10 selling designers with associated category and level:")
print(top_10_selling_designers_with_info)
```

```python
# Count the occurrences of each level
level_counts = df['level'].value_counts()

# Plot the distribution of levels
plt.figure(figsize=(8, 6))
level_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Price Levels')
plt.xlabel('Price Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Consumer preferences in the second-hand market: premium items attract quality-conscious buyers at a discount, cheap items cater to budget shoppers, and average-priced items strike a balance between quality and affordability, appealing broadly.

# Stars Analysis:

```python
# Exploring the distribution of average assigned stars
plt.figure(figsize=(7, 5))
sns.histplot(df['stars'], bins=5, kde=True)
plt.title('Distribution of Average Stars')
plt.xlabel('Average Stars')
plt.ylabel('Frequency')
plt.show()
```

# Platforms Analysis:

```python
# Exploring the distribution by platform
platform_distribution = df['platform'].value_counts().head(3)
print("Platform distribution:\n", platform_distribution)

plt.figure(figsize=(6, 3))
sns.countplot(x='platform', data=df)
plt.title('Platform distribution')
plt.show()
```

Users prefer purchasing from mobile platforms rather than from PCs, and specifically show a greater inclination towards using Android devices rather than Apple.
